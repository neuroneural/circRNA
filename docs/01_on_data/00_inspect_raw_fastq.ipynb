```python
import pandas as pd
import pyfastx
# load a sample fastq file
sample_path = "/Users/ppopov1/_circRNA/RNAseq/01_raw_data/SR006206_20241122/LIB068985_22TLNMLT3_S5_L008_R1_001.fastq.gz"
fa = pyfastx.Fastx(sample_path)

data = []

counter = 0
for item in fa:
    counter += 1
    data.append({
        "id": item[0],
        "seq": item[1],
        "qual": item[2]
    })

    if counter > 1000:
        break

data = pd.DataFrame(data)
data
```
