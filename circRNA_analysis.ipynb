Feature Selection for Sex Prediction Using circRNA
</h2>
<p style="font-family: 'Palatino Linotype', Palatino, serif; color: #555; font-style: italic;">This script predicts sex from circRNA expression data (1,991 features, 27 subjects) using a nested cross-validation framework. The test set is held out entirely until the very end to prevent data leakage. Four feature selection methods — Mutual Information, Recursive Feature Elimination, Random Forest importance, and SHAP values — are each evaluated across 5 outer folds, where feature selection is performed exclusively on each fold's training partition, ensuring that validation samples never influence which features are chosen. Within each outer fold, the selected 50 features are scored using Polyssifier, a multi-classifier benchmarking tool that runs a suite of classifiers under 3-fold inner cross-validation and reports AUC, providing a robust and unbiased comparison across methods. The method with the best mean AUC across outer folds is selected as the winner, and the top 50 features are re-selected on the full training set using that method. Finally, a StandardScaler and RBF-kernel SVM are trained on those features and evaluated on the untouched held-out test set, yielding the final accuracy score.</p>

```python
from circRNA_PreProcess_for_Sex_Prediction import load_circrna_data

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.feature_selection import mutual_info_classif, RFE
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import shap

from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

from polyssifier import poly


# This script predicts sex from circRNA expression data (1,991 features, 27 subjects) using a
# nested cross-validation framework. The test set is held out entirely until the very end to
# prevent data leakage. Four feature selection methods — Mutual Information, Recursive Feature Elimination,
# Random Forest importance, and SHAP values — are each evaluated across 5 outer folds, where
# feature selection is performed exclusively on each fold's training partition, ensuring that validation
# samples never influence which features are chosen. Within each outer fold, the selected 50 features
# are scored using Polyssifier, a multi-classifier benchmarking tool that runs a suite of classifiers
# under 3-fold inner cross-validation and reports AUC, providing a robust and unbiased comparison across
# methods. The method with the best mean AUC across outer folds is selected as the winner, and the top
# 50 features are re-selected on the full training set using that method. Finally, a StandardScaler and
# RBF-kernel SVM are trained on those features and evaluated on the untouched held-out test set,
# yielding the final accuracy score.


# ============================================================
# SETTINGS
# ============================================================

METHODS = ["mi", "rfe", "rf", "shap"]

TOP_N = 50
OUTER_FOLDS = 5
INNER_FOLDS = 3
RANDOM_STATE = 42


# ============================================================
# LOAD DATA
# ============================================================

data = load_circrna_data(scale=False)
X = pd.DataFrame(data["X"])
y = pd.Series(data["y"])

X_train_full, X_test, y_train_full, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=RANDOM_STATE
)


# ============================================================
# FEATURE SELECTION METHODS
# ============================================================

def mi_features(X, y, top_n):
    mi = mutual_info_classif(X, y, random_state=RANDOM_STATE)
    return pd.Series(mi, index=X.columns)\
        .sort_values(ascending=False)\
        .head(top_n)\
        .index.tolist()


def rfe_features(X, y, top_n):
    lr = LogisticRegression(max_iter=5000, solver="saga")
    rfe = RFE(lr, n_features_to_select=top_n, step=0.1)
    rfe.fit(X, y)
    return X.columns[rfe.support_].tolist()


def rf_features(X, y, top_n):
    rf = RandomForestClassifier(
        n_estimators=500,
        random_state=RANDOM_STATE,
        n_jobs=-1
    )
    rf.fit(X, y)
    imp = pd.Series(rf.feature_importances_, index=X.columns)
    return imp.sort_values(ascending=False).head(top_n).index.tolist()



def shap_features(X, y, top_n):

    xgb = XGBClassifier(
        n_estimators=500,
        max_depth=5,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        eval_metric="logloss",
        random_state=RANDOM_STATE,
        n_jobs=-1
    )

    xgb.fit(X, y)

    #explainer = shap.Explainer(xgb.predict,X)
    #shap_values = explainer(X)
    explainer = shap.Explainer(xgb.predict,X)
    shap_values = explainer.shap_values(X)
    importance = np.abs(shap_values).mean(axis=0)
    shap_series = pd.Series(importance,index=X.columns).sort_values(ascending=False)



    return shap_series.head(top_n).index.tolist()




FEATURE_FUNCS = {
    "mi": mi_features,
    "rfe": rfe_features,
    "rf": rf_features,
    "shap": shap_features
}


# ============================================================
# POLY EVALUATION (ONLY EVALUATION ENGINE)
# ============================================================

def evaluate_poly(X_train, y_train, features):

    X_sel = X_train[features].values

    report = poly(
        X_sel,
        y_train,
        n_folds=INNER_FOLDS,
        scale=True,
        scoring="auc",
        project_name="inner_eval"
    )

    try:
        return report["summary"]["mean"]
    except:
        try:
            return float(report)
        except:
            return np.nan


# ============================================================
# OUTER CV: FEATURE SELECTION COMPARISON ONLY
# ============================================================

outer_cv = StratifiedKFold(
    n_splits=OUTER_FOLDS,
    shuffle=True,
    random_state=RANDOM_STATE
)

results = {m: [] for m in METHODS}

print("\n================ FEATURE SELECTION COMPARISON ================\n")

for method in METHODS:

    print(f"\n========== METHOD: {method} ==========\n")

    for fold, (train_idx, val_idx) in enumerate(
        outer_cv.split(X_train_full, y_train_full),
        1
    ):

        X_tr = X_train_full.iloc[train_idx]
        y_tr = y_train_full.iloc[train_idx]

        # -------------------------------
        # FEATURE SELECTION (TRAIN ONLY)
        # -------------------------------
        features = FEATURE_FUNCS[method](X_tr, y_tr, TOP_N)

        # -------------------------------
        # POLYSSIFIER EVALUATION ONLY
        # -------------------------------
        score = evaluate_poly(X_tr, y_tr, features)

        results[method].append(score)

        print(f"Fold {fold} poly AUC: {score:.4f}")


# ============================================================
# SUMMARY
# ============================================================

summary_df = pd.DataFrame({
    "method": METHODS,
    "mean_poly_auc": [np.mean(results[m]) for m in METHODS],
    "std_poly_auc": [np.std(results[m]) for m in METHODS]
})

print("\n================ FINAL SUMMARY ================\n")
print(summary_df)

best_method = summary_df.sort_values(
    "mean_poly_auc",
    ascending=False
)["method"].iloc[0]

print("\nBEST FEATURE METHOD:", best_method)


# ============================================================
# FINAL FEATURE SET (TRAIN FULL)
# ======================================================== ====

print("\n================ FINAL FEATURE SET ================\n")

best_features = FEATURE_FUNCS[best_method](
    X_train_full,
    y_train_full,
    TOP_N
)

for f in best_features:
    print(f)


# ============================================================
# FINAL MODEL: SCALED SVM ONLY
# ============================================================

print("\n================ FINAL SVM MODEL ================\n")

scaler = StandardScaler()

X_train_sel = X_train_full[best_features]
X_test_sel = X_test[best_features]

X_train_scaled = scaler.fit_transform(X_train_sel)
X_test_scaled = scaler.transform(X_test_sel)

final_svm = SVC(
    kernel="rbf",
    probability=True,
    random_state=RANDOM_STATE
)

final_svm.fit(X_train_scaled, y_train_full)

test_score = final_svm.score(X_test_scaled, y_test)

print("\nFINAL TEST ACCURACY (SVM):", test_score)

```
